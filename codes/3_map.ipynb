# Plot of Flats and Points of Interest

This noteboook run in a separate Anaconda environment for Geopanda to work

```python
import pandas as pd
import geopandas as gpd
from geopandas import GeoSeries
from shapely.geometry import Polygon
import matplotlib.pyplot as plt
```

```python
flat = pd.read_csv("../datasets/combined/combined_2015_2018_featured.csv")
gdf_flats = gpd.GeoDataFrame(
    flat, geometry=gpd.points_from_xy(flat.lng, flat.lat))

market = pd.read_csv("../datasets/poi/hawker_market_cleaned.csv")
gdf_market = gpd.GeoDataFrame(
    market, geometry=gpd.points_from_xy(market.lng, market.lat))

mrt = pd.read_csv("../datasets/poi/mrt_cleaned.csv")
gdf_mrt = gpd.GeoDataFrame(
    mrt, geometry=gpd.points_from_xy(mrt.lng, mrt.lat))

park = pd.read_csv("../datasets/poi/park_cleaned.csv")
gdf_park = gpd.GeoDataFrame(
    park, geometry=gpd.points_from_xy(park.lng, park.lat))

school = pd.read_csv("../datasets/poi/school_geocoded.csv")
gdf_school = gpd.GeoDataFrame(
    school, geometry=gpd.points_from_xy(school.lng, school.lat))

mall = pd.read_csv("../datasets/poi/mall_geocoded.csv")
gdf_mall = gpd.GeoDataFrame(
    mall, geometry=gpd.points_from_xy(mall.lng, mall.lat))

sport = pd.read_csv("../datasets/poi/sport_cleaned.csv")
gdf_sport = gpd.GeoDataFrame(
    sport, geometry=gpd.points_from_xy(sport.lng, sport.lat))

attract = pd.read_csv("../datasets/poi/attraction_cleaned.csv")
gdf_attract = gpd.GeoDataFrame(
    attract, geometry=gpd.points_from_xy(attract.lng, attract.lat))
```

```python
# load the geojson file downloaded from data.gov.sg

df = gpd.read_file('../datasets/input/national-map-polygon-geojson.geojson')

ax = df.plot(figsize=(30, 15), alpha=0.5, edgecolor='k')

plt.rcParams['axes.linewidth'] = 0.1 #set the value globally

# We can now plot the points from ``GeoDataFrame``.
gdf_flats.plot(ax=ax, color='blue',markersize=200, label='HDB Flats')
gdf_park.plot(ax=ax, color='green', markersize=200,label='Parks')
gdf_market.plot(ax=ax, color='red', markersize=100,label='Hawker Markets')
gdf_school.plot(ax=ax, color='magenta',markersize=100,label='Schools')
gdf_mall.plot(ax=ax, color='yellow',markersize=100,label='Malls')
gdf_sport.plot(ax=ax, color='#c6c8cb',markersize=100,label='Sports Facilities')
gdf_attract.plot(ax=ax, color='#61371F',markersize=100,label='Attractions')
gdf_mrt.plot(ax=ax, color='black',marker='X',markersize=100,label='MRT stations')

plt.title('Locations of HDB flats and Points of Interest',fontsize=15)
plt.legend()
df.plot;
```

There are more than a million HDB flats spread over 24 towns in Singapore.

"In developing towns and estates, HDB takes a comprehensive approach where residents’ needs and lifestyle aspirations play an important part in the planning.
Our towns are designed to be self-sufficient, with easy access to shops, schools, and social and recreational facilities, and an abundance of greenery"[source: HDB](https://www.hdb.gov.sg/cs/infoweb/about-us/history/town-planning)

In the above map you can see that MRT stations, malls, market, schools, sport facitlities, libraries and nature parks are close to HDB flats. Those living in downtown core have the added advantage of being close to most attractions.
